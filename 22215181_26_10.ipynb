Generar una matriz a que tenga 300 x 300

que tenga los numeros aleatorios desde 0 a los 3 primeros numeros de mi numero de expediente 

```python
import random

# Tamao de la matriz
filas = 300
columnas = 300

# Crear una matriz de 300x300 con nmeros aleatorios
matriz = [[random.randint(0, 222) for _ in range(columnas)] for _ in range(filas)]

# Imprimir la matriz
for fila in matriz:
    print(' '.join(map(str, fila)))
```
